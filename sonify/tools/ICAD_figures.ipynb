```python
case = 'c0029'
Pxx_scaled  = np.load(base_path.joinpath('data','aneurisk','spectrograms',case,'sac_averaged','sac_spectro_unfiltered_scaled.npy'))
bins 		= np.load(base_path.joinpath('data','aneurisk','spectrograms',case,'sac_averaged','bins.npy'))
freqs 		= np.load(base_path.joinpath('data','aneurisk','spectrograms',case,'sac_averaged','freqs.npy'))	
```

```python
tonal_synth = spectro_harmonics(Pxx_scaled,freqs,bins)
tonal_synth.compute_features()
```

```python
fig,ax = plt.subplots(1,3,figsize=(15,3))

gs1 = gridspec.GridSpec(7, 3)
gs1.update(wspace=0.025, hspace=0.05) # set the spacing between axes. 

plt.tight_layout()
fig.tight_layout()

ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')

ax[0].pcolormesh(bins,freqs,Pxx_scaled,shading='gouraud')

ax[1].pcolormesh(bins,freqs,Pxx_scaled,shading='gouraud')
ax[1].plot(bins, tonal_synth.noise,color='tab:olive',linewidth=4)
ax[1].fill_between(bins,tonal_synth.noise,color='bisque',alpha=0.15)

#ax[1].fill_between(bins,tonal_synth.noise,color='darkslateblue')
ax[2].plot(bins,tonal_synth.noise,color='tab:olive',linewidth=4)
ax[2].pcolormesh(bins,freqs,np.where(tonal_synth.labels>0,1,0),cmap='viridis',shading = 'gouraud')
ax[2].fill_between(bins,tonal_synth.noise,color='bisque',alpha=0.15)

plt.subplots_adjust(wspace=0.02, hspace=0,top=1,bottom=0,right=1,left=0)

plt.savefig('/home/lucas/Documents/viz/renders/ICAD2021/Figures/Figure4/figure4.png',dpi=500)
```
